```python
def get_match_info(venue_name, team_name):
  """
  This function takes a DataFrame, venue name and team name as input and returns number of matches played at the venue, 
  result of the match, man of the match, highest run scorer, and most wicket taker bowler for the given team

  Args:
      venue_name: The name of the venue to filter data
      team_name: The team name to filter data
      data: A Pandas DataFrame containing match data

  Returns:
      A dictionary containing the following information:
          matches_played: The number of matches played at the venue for the team
          result: The result of the last match played at the venue for the team
          man_of_the_match: The man of the match for the last match played at the venue for the team
          highest_run_scorer: The highest run scorer for the team in the last match played at the venue
          most_wicket_taker: The most wicket taker bowler for the team in the last match played at the venue
  """
  filtered_data = tba_df.loc[(tba_df['venue'] == venue_name) & (tba_df['winner'] == team_name)]
  matches_played = len(filtered_data)
  
  if matches_played > 0:
    last_match_data = filtered_data.iloc[0]
    result = last_match_data['result']
    man_of_the_match = last_match_data['player_of_match']
    highest_run_scorer = filtered_data[filtered_data['batsman_runs'] == filtered_data['batsman_runs'].max()]['batsman'].iloc[0]
    most_wicket_taker = filtered_data[filtered_data['is_wicket'] == 1]['bowler'].value_counts().idxmax()
  else:
    result = "No matches played"
    man_of_the_match = "NA"
    highest_run_scorer = "NA"
    most_wicket_taker = "NA"
  
  return {
      "matches_played": matches_played,
      "result": result,
      "man_of_the_match": man_of_the_match,
      "highest_run_scorer": highest_run_scorer,
      "most_wicket_taker": most_wicket_taker
  }
```

```python
outliers =[]
def dectec_outliers(data):
    threshold=3
    mean= np.mean(data)
    std= np.std(data)
    for i in data:
        z=(i-mean)/std
        if np.abs(z)> threshold:
            outliers.append(i)
    return outliers
```
